```python
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('metadata_stepfreq.csv')
```

```python
# Any interesting between weight and step_freq?

plt.subplot(3, 1, 1)
plt.plot(data['Weight'], data['x_freq'], 'g.')
plt.subplot(3, 1, 2)
plt.plot(data['Weight'], data['y_freq'], 'b.')
plt.subplot(3, 1, 3)
plt.plot(data['Weight'], data['z_freq'], 'r.')
plt.show()

# Doesn't look like there's a pattern here.
```

```python
# Any interesting between height and step_freq?

plt.subplot(3, 1, 1)
plt.plot(data['Height (CM)'], data['x_freq'], 'g.')
plt.subplot(3, 1, 2)
plt.plot(data['Height (CM)'], data['y_freq'], 'b.')
plt.subplot(3, 1, 3)
plt.plot(data['Height (CM)'], data['z_freq'], 'r.')
plt.show()

# Also doesn't look like there's a pattern here...
```

```python
# BMI??

data['BMI'] = data['Weight'] / ((data['Height (CM)'] / 100)**2)

plt.subplot(3, 1, 1)
plt.plot(data['BMI'], data['x_freq'], 'g.')
plt.subplot(3, 1, 2)
plt.plot(data['BMI'], data['y_freq'], 'b.')
plt.subplot(3, 1, 3)
plt.plot(data['BMI'], data['z_freq'], 'r.')
plt.show()

# Nope
```

```python
# what about age?

plt.subplot(3, 1, 1)
plt.plot(data['Age'], data['x_freq'], 'g.')
plt.subplot(3, 1, 2)
plt.plot(data['Age'], data['y_freq'], 'b.')
plt.subplot(3, 1, 3)
plt.plot(data['Age'], data['z_freq'], 'r.')
plt.show()

# wow, nothing
```
